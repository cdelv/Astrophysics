Definicón de una función que recibe posición y velocidad de N partículas junto con sus masas


```python
def energy(state,mass):
  G=6.67408e-11
  T=0
  L=0
  V=0
  for j in range(len(state)):
    T+=state[j][3]**2+state[j][4]**2+state[j][5]**2
  for j in range(len(state)):
    for i in range(j+1,len(state)):
      V+=-G*mass[j]*mass[i]/((state[j][0]-state[i][0])**2+(state[j][1]-state[i][1])**2+(state[j][2]-state[i][2])**2)
  E=T+V
  for j in range(len(state)):
    r=[state[j][0],state[j][1],state[j][2]]
    P=[state[j][3]*mass[j],state[j][4]*mass[j],state[j][5]*mass[j]]
    L+=np.cross(r,P)
  return [E,L]
```

En este caso la energía total es la energía cinética más la energía potencial de las dos partículas
\begin{equation}
  E=T+V=\sum_{i=1}^N\frac{1}{2}m_iv_i^2+\sum_{i=1}^N\sum_{j>i}^N\frac{-Gm_im_j}{|r_{ij}|}
\end{equation}

El calculo del momento angular se hace de la siguiete manera
\begin{equation}
L=\sum_{i=1}^NL_i
\end{equation}
\begin{equation}
L_i=r_i\times P_i
\end{equation}
donde $r_i$ es el vector posición y $P_i$ es el momentum de la partícula iésima.
